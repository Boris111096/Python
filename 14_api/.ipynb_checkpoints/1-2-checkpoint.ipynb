```python
req = requests.get('https://habr.com/ru/all/')
soup = BeautifulSoup(req.text, 'html.parser')
posts = soup.find_all('article', class_='post')
```

### Задание 2.
#### Обязательная часть
Написать скрипт, который будет проверять список e-mail адресов на утечку при помощи сервиса Avast Hack Ckeck. Список email-ов задаем переменной в начале кода:
EMAIL = [xxx@x.ru, yyy@y.com]

В итоге должен формироваться датафрейм со столбцами: - <дата утечки> - <источник утечки> - <описание утечки>.

Подсказка: сервис работает при помощи "скрытого" API. Внимательно изучите post-запросы.

```python
EMAIL = ['xxx@x.ru', 'yyy@y.com']

URL = 'https://identityprotection.avast.com/v1/web/query/breaches-stats'

headers = { 'Accept': 'application/json',
            'Accept-Language': 'ru-RU',
            'Connection': 'keep-alive',
            'Content-Length': '0',
            'Host': 'identityprotection.avast.com',
            'Origin': 'https://www.avast.com',
            'Referer': 'https://www.avast.com/hackcheck/',
            'Sec-Fetch-Dest': 'empty',
            'Sec-Fetch-Mode': 'cors',
            'Sec-Fetch-Site': 'same-site',
            'User-Agent': 'Mozilla/5.0 (x64)',
            'Vaar-Header-App-Product': 'hackcheck-web-avast',
            'Vaar-Version': '0'
          }

df = pd.DataFrame()

for email in EMAIL:
    res = requests.post(URL, data = email, headers = headers)
    for leak in res.json()[ 'biggest']:
        df = pd.concat([df, pd.DataFrame([{ 'email'         : email,
                                            'date_leaked'   : leak['publishDate'],
                                            'source_leaked' : leak['site'],
                                            'desk_leaked'   : leak['description']
                                                      }])])
df
```
