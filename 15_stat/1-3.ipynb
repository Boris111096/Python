### Задание 2. Работа с выбросами

В выбранных числовых столбцах найти выбросы, выдвинуть гипотезы об их причинах и проинтерпретировать результаты. Принять и обосновать решение о дальнейшей работе с ними.

```python
new_df = my_df
new_df.describe(include = 'all')
```

Выбросы присутствуют в стобцах:
* rectal_temperature
* pulse
* total_protein  


1. rectal_temperature:  
    Тут выбросы скорее всего связаны с погрешностью измерений. Можно попробовать найти выбросы через межквартальный размах и отсечь слишком высокие / слишком низкий результаты.
2. pulse:  
    Присутствуют очень высокие значения. Возможно человеческий фактор при записи / измерении, либо же какие-то аномальные значения пульса, которые в любом случае в статистике ни к чему. Делаем то же, что в п.1.
3. total_protein:  
    То же, что в п.2.

```python
def find_outlier(df, key):
    q1 = df[key].quantile(0.25)
    q3 = df[key].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr) 
    upper_bound = q3 + (1.5 * iqr)
    remove_outliers = df[df[key].between(lower_bound, upper_bound, inclusive=True)].sort_values(str(key))
    return remove_outliers
```

```python
new_df = find_outlier(new_df, 'rectal_temperature')
new_df
```

```python
new_df = find_outlier(new_df, 'pulse')
new_df
```

```python
new_df = find_outlier(new_df, 'total_protein')
new_df
```
