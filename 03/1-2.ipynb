# Задание 1

```python
# Основная программа
# Я разбил функции по отдельным ячейкам, но в самом низу есть ячейка, включающая все функции вместе

documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

def requests():
    request = 0
    while request != 'q':
        request = input('Введите команду: ')
        if request == 'q':
            print('End')
        elif request == 'p':
            doc_owner()
        elif request == 's':
            where()
        elif request == 'l':
            info()
        elif request == 'ads':
            add()
        elif request == 'ds':
            delete()
        elif request == 'ad':
            add_doc()
        elif request == 'd':
            del_doc()
        elif request == 'm':
            move()
        else:
            print('Нет такой команды')

    return

requests()
```

```python
# Пункт 1
# Пользователь по команде "p" может узнать владельца документа по его номеру

def doc_owner():
    doc_number = input('Введите номер документа: ')
    for document in documents:
        if doc_number in document.values():
            return print('Владелец документа – ',document.get('name'))
    return print('Нет такого документа')
doc_owner()
```

```python
# Пункт 2
# Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится

def where():
    doc_number = input('Введите номер документа: ')
    for directory in directories:
        if doc_number in directories.get(str(directory)):
            return print('Документ хранится на полке: ',directory)
    return print('Нет такого документа')
where()
```

```python
# Пункт 3
# Пользователь по команде "l" может увидеть полную информацию по всем документам

def info():
    for document in documents:
        print('\n№',document['number'],'\nТип:',document['type'],'\nИмя:',document['name'])
        for directory in directories:
            if document['number'] in directories.get(directory):
                print('Полка:',directory,'\n')
info()
```

```python
# Пункт 4
# Пользователь по команде "ads" может добавить новую полку

def add():
    shelf_num = input('Введите номер полки для добавления: ')
    if directories.get(shelf_num) != None:
        print('Такая полка уже существует. Текущий перечень полок: ',*list(directories.keys()))
    else:
        directories[shelf_num] = []
        print('Текущий перечень полок: ',*list(directories.keys()))
add()
```

```python
# Пункт 5
# Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)

def delete():
    shelf_num = input('Введите номер полки: ')
    if directories.get(shelf_num) == None:
        print('Такой полки не существует. Текущий перечень полок: ',*list(directories.keys()))
    elif directories.get(shelf_num) == []:
        del(directories[shelf_num])
        print('Полка удалена. Текущий перечень полок: ',*list(directories.keys()))
    else:
        print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ',*list(directories.keys()))

delete()
```

# Задание 2 

```python
# Пункт 1.
# Пользователь по команде "ad" может добавить новый документ в данные

def add_doc():
    type_ = input('Введите тип документа: ')
    number = input('Введите номер документа: ')
    name = input('Введите владельца документа: ')
    shelf = input('Введите полку для хранения: ')
    if shelf not in directories.keys():
        return print('Такой полки не существует. Добавьте полку командой as.')
    documents.append({'type': type_, 'number': number, 'name': name}) 
    directories[shelf] = number
    print('Документ добавлен. Текущий список документов:')
    info()
    
add_doc()
```

```python
# Пункт 2
# Пользователь по команде "d" может удалить документ из данных

def del_doc():
    doc = input('Введите номер документа: ')
    for document in documents:
        if doc == document.get('number'):
            del(documents[documents.index(document)])
            print('Документ удален. Текущий список документов:')
            info()
            return 
    print('Документ не найден в базе. Текущий список документов:')
    info()
    return
del_doc()
```

```python
# Пункт 3
# Пользователь по команде "m" может переместить документ с полки на полку

def move():
    doc = input('Введите номер документа: ')
    shelf = input('Введите номер полки: ')
    if shelf not in directories.keys():
        print('Такой полки не существует. Текущий список документов:')
        info()
        return 
    for directory in directories:
        if doc in directories.get(directory):
            directories[shelf].append(doc)
            directories[directory].remove(doc)
            print('Документ перемещен. Текущий список документов:')
            info()
            return 
    print('Документ не найден в базе. Текущий список документов:')
    info()
    return
            
move()
```
