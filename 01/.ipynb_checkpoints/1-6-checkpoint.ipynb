```python
# Задание 6

shape = str(input("Назовите фигуру: "))

if shape == 'Треугольник':
    a = int(input("1 сторона: "))
    b = int(input("2 сторона: "))
    c = int(input("3 сторона: "))
    p = (a+b+c)/2
    s = (p*(p-a)*(p-b)*(p-c))**(1/2)
    print('Площадь = ',s)
elif shape == 'Круг':
    r = int(input("Радиус круга: "))
    import math
    s = math.pi*r**2
    print('Площадь = ',s)
elif shape == 'Прямоугольник':
    a = int(input("1 сторона: "))
    b = int(input("2 сторона: "))
    s = a*b
    print('Площадь = ',s)
else:
    print ('Неверное название')
```

```python
# Задание 2 (Убрал пробелы.
#             Сделал сначала по той статье, что вы скинули, мол:
#             "Если год делится как на 4, так и на 100, возможно, это не високосный год."
#             И тогда нужно проверить:
#             "Если год ДЕЛИТСЯ на 100 и НЕ ДЕЛИТСЯ на 400, это НЕ ВИСОКОСНЫЙ год."
#             По такой локике получается, что, например, 300 год - НЕ високосный. Но это же не правда. Можно даже
#             простоза гуглить этот год и узнать, что он был високосным.
#             Короче такой "скрипт" не работает. Проверки деления на 4 вполне достаточно, можете ниже вот убедиться:

year = 0
q1 = 0
q2 = 0
while year < 2020:
    year += 1
    if year / 4 == year // 4:
        print(year,'год - Високосный')
        q1 += 1
    else:
        print(year,'год - Обычный')
        q2 += 1
print('___________________________________')
print(q1,' - количество високосных лет')
print(q2,' - количество обычных лет')
q = q1 + q2
print(q1, ' + ', q2, ' = ', q, ' — всего — (' ,q1 / q * 100, '% високосных лет /', q2 / q * 100, '% обычных лет)')
```

```python
# Задание 6 (Убрал / поставил пробелы. Убрал str у пользовательского ввода)

shape = input("Назовите фигуру: ") 

if shape == 'Треугольник':
    a = int(input("1 сторона: "))
    b = int(input("2 сторона: "))
    c = int(input("3 сторона: "))
    p = ( a + b + c) / 2
    s = (p * (p - a) * (p - b) * (p - c)) ** (1 / 2)
    print('Площадь = ', s)
elif shape == 'Круг':
    r = int(input("Радиус круга: "))
    import math
    s = math.pi * r ** 2
    print('Площадь = ',s)
elif shape == 'Прямоугольник':
    a = int(input("1 сторона: "))
    b = int(input("2 сторона: "))
    s = a * b
    print('Площадь = ', s)
else:
    print('Неверное название')
```
